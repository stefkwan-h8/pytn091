# Principal Component Analysis
tipe unsupervised learning

bertujuan mengurangi dimensi sebuah data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# kita gunakan contoh data breast cancer dataset dari sklearn
from sklearn.datasets import load_breast_cancer

# kita import juga machine learning PCA dari sklearn
from sklearn.decomposition import PCA

# untuk scaling data
from sklearn.preprocessing import StandardScaler
```

```python
# 1. load, import data
data_breast_cancer = load_breast_cancer()

col_names = data_breast_cancer.feature_names
# print(col_names)

df_breast_cancer = pd.DataFrame(data_breast_cancer.data, columns=col_names)
df_breast_cancer
```

```python
df_breast_cancer.info()
```

```python
target_breast_cancer = data_breast_cancer.target

print(data_breast_cancer.target_names)

target_breast_cancer

df_breast_cancer_with_target = df_breast_cancer.copy()
df_breast_cancer_with_target["label"] = target_breast_cancer
df_breast_cancer_with_target.head()

# ini data dengan tambahan 1 kolom target
# untuk contoh PCA machine learning, kita tidak gunakan targetnya, jadi tidak kumasukkan ke df_breast_cancer
```

```python
# 2. explore data, prep data
```

data sudah rapi, tidak ada missing value, data types udah numerik semua, ... 

untuk PCA kita perlu scale datanya

```python
scaler = StandardScaler()
df_breast_cancer_scaled = pd.DataFrame(scaler.fit_transform(df_breast_cancer), columns=col_names)

df_breast_cancer_scaled
```

```python
# 3. membuat dan melatih machine learning

model = PCA(n_components=2) # pilih n_components = tarik berapa banyak Principal Component
hasil = model.fit_transform(df_breast_cancer_scaled)

col_names_PC = []
for n in range(len(hasil[0])):
    col_names_PC.append("PC" + str(n + 1))

pd.DataFrame(hasil, columns=col_names_PC)
```

```python
pd.DataFrame(model.components_, index=col_names_PC, columns=col_names)
```

setiap Principal Component merupakan sebuah axis baru (1 dimensi baru) yang terbuat dari campuran seluruh fitur data awal yang kita berikan.

Di contoh ini, kita menggunakan 30 fitur - ada 30 dimensi di data awal kita - dan setiap Principal Component terbuat dari campuran 30 fitur ini.

PC1 terbuat dari 0.2 mean radius + 0.1 mean texture + ...

sebenarnya bisa saja kita menggunakan jumlah dimensi yang sama dan membuat PC sebanyak jumlah fitur awal, tapi ini tidak mendapat hasil yang kita inginkan saat menggunakan PCA, yaitu mengurangi dimensi data.

```python
pd.DataFrame(model.explained_variance_ratio_, index=col_names_PC, columns=["Explained Variance Ratio"])
```

```python
model.explained_variance_ratio_.sum()
```

jika di data ini kita menggunakan 2 principal component saja, kita hanya bisa menjelaskan 63% penyebaran data yang ada. Ada penyebaran data yang tidak ter capture di dua dimensi PC1 dan PC2
